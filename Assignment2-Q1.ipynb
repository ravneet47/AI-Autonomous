# Q1. Implement K-mean cluster algorithm using the same data in the lecture

Your implementation should include:
- `Two clusters`

- `Three clusters`

- `Four clusters`

- `Plot the data points for each cluster for each iteration`

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
```

```python
data = np.array([
    [7, 7], [5, 9], [6, 9], [7, 9], [5, 11], [5, 3], [6, 1], [6, 2], [7, 1], 
    [7, 2], [7, 3], [8, 4], [8, 6], [9, 3], [9, 4], [9, 5], [10, 4], [10, 5], 
    [10, 6], [9, 7]
])
```

```python

def kmeans(data, k, max_iters=10):
    np.random.seed(42)
    centroids = data[np.random.choice(data.shape[0], k, replace=False)]
    for i in range(max_iters):
        
        # Assign clusters
        distances = np.linalg.norm(data[:, np.newaxis] - centroids, axis=2)
        labels = np.argmin(distances, axis=1)

        # Plotting
        plt.figure()
        plt.scatter(data[:, 0], data[:, 1], c=labels, s=30, cmap='viridis')
        plt.scatter(centroids[:, 0], centroids[:, 1], c='red', s=200, alpha=0.75)
        plt.title(f'Iteration {i+1}')
        plt.show()

        # Recalculate centroids
        new_centroids = np.array([data[labels == j].mean(axis=0) for j in range(k)])
        if np.all(centroids == new_centroids):
            break
        centroids = new_centroids

    return centroids, labels

# Function to run kmeans for different k values
def run_kmeans(data, k_values):
    for k in k_values:
        print(f"\nRunning K-means for k={k}")
        centroids, labels = kmeans(data, k)
```

## K = 2

```python

run_kmeans(data, k_values=[2])
```

## K = 3

```python
run_kmeans(data, k_values=[3])
```

## K = 4

```python
run_kmeans(data, k_values=[4])
```
